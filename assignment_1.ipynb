```python
# Using the data from assignment 1, build a model predicting the Appeal of a report based on its attribute scores
```

```python
df_reviews.head(2)
```

```python
X = df_reviews.drop(['ReviewId', 'ReportId', 'UserId', 'FamiliarityWithBrand', 'Appeal'], axis=1)
y = df_reviews['Appeal']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train:', y_train.shape)
print('y_test:', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg = LinearRegression()

linear_reg.fit(X_train, y_train)
y_pred = linear_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
print(r2_score)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rand_forst= RandomForestRegressor()
```

```python
rand_forest_model = rand_forst.fit(X_train, y_train)
y_pred = rand_forst.predict(X_test)
```

```python
import numpy as np
from sklearn import metrics
```

```python
print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error (RMSE):', metrics.mean_squared_error(y_test, y_pred, squared=False))
```
